```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/Retail_Transactions_Dataset.csv')

df.head()
```

```python
total_itens = df['Total_Items'].sum()
faturamento_total = df['Total_Cost'].sum()

print(f"Total de itens vendidos: {total_itens}")
print(f"Faturamento Total: R$ {faturamento_total:.2f}")
```

```python
vendas_por_loja = df.groupby('Store_Type')['Total_Items'].sum().reset_index()
vendas_por_loja['Percentual'] = (vendas_por_loja['Total_Items'] / total_itens) * 100

print(vendas_por_loja)
```

```python
custo_por_loja = df.groupby('Store_Type')['Total_Cost'].sum().reset_index()
faturamento_global = df['Total_Cost'].sum()
custo_por_loja['% do Custo Total'] = (custo_por_loja['Total_Cost'] / faturamento_global) * 100

print("--- Custos por Tipo de Loja ---")
print(custo_por_loja)
```

```python
pagamentos = df['Payment_Method'].value_counts()

print("\n--- Uso dos Métodos de Pagamento ---")
print(pagamentos)

print(f"\nO método MAIS usado é: {pagamentos.idxmax()}")
print(f"O método MENOS usado é: {pagamentos.idxmin()}")
```

```python
media_custo = df['Total_Cost'].mean()
# filtrando somente as compras acima da média e vendo os métodos
compras_caras = df[df['Total_Cost'] > media_custo]['Payment_Method'].value_counts()

print(f"Média de custo: R$ {media_custo:.2f}")
print("\nMétodos usados em compras acima da média:")
print(compras_caras)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Hora'] = df['Date'].dt.hour
df['Dia_Semana'] = df['Date'].dt.day_name()

vendas_hora = df['Hora'].value_counts().sort_index()

vendas_dia = df['Dia_Semana'].value_counts()

print("--- Vendas por Hora ---")
print(vendas_hora)

print("\n--- Vendas por Dia da Semana ---")
print(vendas_dia)
```

```python
# tabela dinâmica p ver o volume de vendas
# Linhas: Cidade | Colunas: Hora | Valores: Contagem de transações
vendas_cidade_hora = df.pivot_table(index='City', columns='Hora', values='Transaction_ID', aggfunc='count')

vendas_cidade_hora.head()
```

```python
import pandas as pd

# 1. Garante que a coluna Date seja tratada como data pelo Python
df['Date'] = pd.to_datetime(df['Date'])

df['Day'] = df['Date'].dt.day_name()  
df['Hour'] = df['Date'].dt.hour      

vendas_agrupadas = df.groupby(['City', 'Day', 'Hour'])['Transaction_ID'].count()

pico_geral = vendas_agrupadas.sort_values(ascending=False).index[0]
valor_pico = vendas_agrupadas.max()

print(f"A maior concentração de vendas ocorreu em:")
print(f"Cidade: {pico_geral[0]}")
print(f"Dia: {pico_geral[1]}")
print(f"Horário: {pico_geral[2]}h")
print(f"Total de transações: {valor_pico}")
```

```python
vendas_por_grupo = df.groupby(['City', 'Day', 'Hour'])['Transaction_ID'].count()

picos_cidades = vendas_por_grupo.groupby('City').idxmax()

print("--- Dia e Hora com mais vendas por Cidade ---")
for cidade, info in picos_cidades.items():
    qtd = vendas_por_grupo[info]
    print(f"Cidade: {info[0]} | Dia: {info[1]} | Horário: {info[2]}h | Total: {qtd} vendas")
```

```python
# nv dataFrame com convenience store
df_conv = df[df['Store_Type'] == 'Convenience Store'].copy()
print(f"Total de vendas no setor de Conveniência: {len(df_conv)}")
```

```python

#trocando [ e ] e ' por nada
df_conv['Produto_Limpo'] = df_conv['Product'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", "", regex=False)

df_conv['Product_List'] = df_conv['Produto_Limpo'].str.split(', ')

df_itens_separados = df_conv.explode('Product_List')

ranking_itens = df_itens_separados['Product_List'].value_counts()

print("--- Ranking CORRETO de itens na Conveniência ---")
print(ranking_itens.head(10))
```

```python
item_mais_vendido = ranking_itens.idxmax()
item_menos_vendido = ranking_itens.idxmin()

print(f"O item campeão de vendas é: {item_mais_vendido}")
print(f"O item que menos saiu é: {item_menos_vendido}")
```

```python
perfil_compra = df_itens_separados.groupby('Customer_Category')['Product_List'].agg(lambda x: x.value_counts().idxmax())

print("--- Item favorito por perfil de cliente ---")
print(perfil_compra)
```

No entanto, devido à alta dimensionalidade do dataset de conveniência (mais de 160.000 colunas após o encoding), o sistema apresentou um erro de memória (MemoryError). Esse algoritmo exige a geração prévia de todos os conjuntos de itens candidatos, o que se tornou inviável para o hardware disponível.

Para solucionar esse gargalo, optou-se pela migração para o algoritmo FP-Growth (Frequent Pattern Growth). Ao contrário do Apriori, o FP-Growth utiliza uma estrutura de árvore compacta para armazenar as frequências, o que reduz drasticamente o consumo de RAM e acelera o processamento em bases de dados de varejo pulverizadas.

O que isso significa para o setor de Conveniência?

    Independência de Compra: Os dados sugerem que as compras neste setor são predominantemente pontuais e independentes. Diferente de um supermercado, onde há um planejamento de "compra de mês", na conveniência o cliente busca suprir uma necessidade imediata e específica (ex: comprar apenas a pasta de dente).

    Pulverização do Mix: Com um ranking de vendas liderado de forma isolada pela Pasta de Dente (Toothpaste), não há uma "âncora" secundária que arraste outros produtos com frequência estatística relevante.

Sugestão Estratégica:
Recomenda-se focar em estratégias de exposição de produtos (cross-merchandising) baseadas no tráfego e não apenas em afinidade lógica, já que o comportamento do consumidor é de conveniência pura.
