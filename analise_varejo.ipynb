```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# carregando dataset
df = pd.read_csv('Data/Retail_Transactions_Dataset.csv')

# visualizando as primeiras 5 linhas 
df.head()
```

```python
total_itens = df['Total_Items'].sum()
faturamento_total = df['Total_Cost'].sum()

print(f"Total de itens vendidos: {total_itens}")
print(f"Faturamento Total: R$ {faturamento_total:.2f}")
```

```python
vendas_por_loja = df.groupby('Store_Type')['Total_Items'].sum().reset_index()
vendas_por_loja['Percentual'] = (vendas_por_loja['Total_Items'] / total_itens) * 100

print(vendas_por_loja)
```

```python
custo_por_loja = df.groupby('Store_Type')['Total_Cost'].sum().reset_index()
faturamento_global = df['Total_Cost'].sum()
custo_por_loja['% do Custo Total'] = (custo_por_loja['Total_Cost'] / faturamento_global) * 100

print("--- Custos por Tipo de Loja ---")
print(custo_por_loja)
```

```python
pagamentos = df['Payment_Method'].value_counts()

print("\n--- Uso dos Métodos de Pagamento ---")
print(pagamentos)

print(f"\nO método MAIS usado é: {pagamentos.idxmax()}")
print(f"O método MENOS usado é: {pagamentos.idxmin()}")
```

```python
media_custo = df['Total_Cost'].mean()
# filtrando somente as compras acima da média e vendo os métodos
compras_caras = df[df['Total_Cost'] > media_custo]['Payment_Method'].value_counts()

print(f"Média de custo: R$ {media_custo:.2f}")
print("\nMétodos usados em compras acima da média:")
print(compras_caras)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Hora'] = df['Date'].dt.hour
df['Dia_Semana'] = df['Date'].dt.day_name()

# Vendas por hora (quais horários o pessoal mais compra?)
vendas_hora = df['Hora'].value_counts().sort_index()

# Vendas por dia da semana
vendas_dia = df['Dia_Semana'].value_counts()

print("--- Vendas por Hora ---")
print(vendas_hora)

print("\n--- Vendas por Dia da Semana ---")
print(vendas_dia)
```

```python
# Criando uma tabela dinâmica para ver o volume de vendas
# Linhas: Cidade | Colunas: Hora | Valores: Contagem de transações
vendas_cidade_hora = df.pivot_table(index='City', columns='Hora', values='Transaction_ID', aggfunc='count')

# Mostrando as primeiras cidades para conferir
vendas_cidade_hora.head()
```
